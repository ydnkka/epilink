# EpiLink interactive heatmap walkthrough

This notebook focuses on one interactive view: a compatibility heatmap across ancestor-descendant and common-ancestor scenario pairs.

The heatmap uses:

- x-axis: ancestor-descendant scenarios `ad(m)`
- y-axis: common-ancestor scenarios `ca(m_i,m_j)`

Use the sliders to evaluate each `[ad(m), ca(m_i,m_j)]` combination at a chosen sample-time difference and genetic distance.

Run this notebook in the `epilink` conda environment so that `epilink`, `scipy`, and `ipywidgets` are available.

```python
import matplotlib.pyplot as plt
import numpy as np

from epilink import EpiLink, InfectiousnessToTransmission, NaturalHistoryParameters

plt.style.use('ggplot')

rng_seed = 2026
```

## Build an EpiLink model

We use the infectiousness-to-transmission profile and precompute Monte Carlo draws for a modest hidden-depth limit.

```python
parameters = NaturalHistoryParameters()
profile = InfectiousnessToTransmission(parameters=parameters, rng_seed=rng_seed)

epilink = EpiLink(
    transmission_profile=profile,
    maximum_depth=5,
    mc_samples=8000,
    mutation_process="stochastic",
)
```

## Interactive compatibility heatmap
The figure combines single-scenario compatibility profiles for $\text{ad}(m)$ and $\text{ca}(m_i,m_j)$, a ranked view of the strongest scenario pairs, and a heatmap of the two-scenario subsets $[\text{ad}(m), \text{ca}(m_i,m_j)]$.

The controls now set the observed values as positive integers:

- sample-time difference $t_{i,j}$
- genetic distance $g_{i,j}$

```python
import ipywidgets as widgets
from IPython.display import display

ad_labels = [scenario.label() for scenario in epilink.scenarios if scenario.kind == "ad"]
ca_labels = [scenario.label() for scenario in epilink.scenarios if scenario.kind == "ca"]
ad_models = {label: epilink.pairwise_model(label) for label in ad_labels}
ca_models = {label: epilink.pairwise_model(label) for label in ca_labels}

combination_models = {
    (ca_label, ad_label): epilink.pairwise_model([ad_label, ca_label])
    for ca_label in ca_labels
    for ad_label in ad_labels
}

all_time_draws = np.concatenate(
    [draws["time_draws"] for draws in epilink.draws_by_scenario.values()]
)
all_genetic_draws = np.concatenate(
    [draws["genetic_draws"] for draws in epilink.draws_by_scenario.values()]
)

def integer_slider_config(values):
    lower = 0
    upper = max(lower, int(np.ceil(np.percentile(values, 99))))
    initial = int(np.clip(np.rint(np.median(values)), lower, upper))
    return initial, lower, upper


time_default, time_min, time_max = integer_slider_config(all_time_draws)
genetic_default, genetic_min, genetic_max = integer_slider_config(
    all_genetic_draws
)

time_slider = widgets.IntSlider(
    value=time_default,
    min=time_min,
    max=time_max,
    description="$t_{i,j}$:",
    continuous_update=False,
    layout=widgets.Layout(width="520px"),
)

genetic_slider = widgets.IntSlider(
    value=genetic_default,
    min=genetic_min,
    max=genetic_max,
    description="$g_{i,j}$:",
    continuous_update=False,
    layout=widgets.Layout(width="520px"),
)

output = widgets.Output()


def compatibility_profile(models, labels, sample_time_difference, genetic_distance):
    return np.array(
        [
            float(models[label](sample_time_difference, genetic_distance))
            for label in labels
        ],
        dtype=float,
    )


def compatibility_matrix(sample_time_difference, genetic_distance):
    return np.array(
        [
            [
                float(
                    combination_models[(ca_label, ad_label)](
                        sample_time_difference,
                        genetic_distance,
                    )
                )
                for ad_label in ad_labels
            ]
            for ca_label in ca_labels
        ],
        dtype=float,
    )


def ranked_combinations(matrix):
    return sorted(
        [
            (matrix[row_index, col_index], ca_labels[row_index], ad_labels[col_index])
            for row_index in range(len(ca_labels))
            for col_index in range(len(ad_labels))
        ],
        key=lambda item: item[0],
        reverse=True,
    )


def draw_heatmap(*_):
    sample_time_difference = int(time_slider.value)
    genetic_distance = int(genetic_slider.value)
    ad_profile = compatibility_profile(
        ad_models,
        ad_labels,
        sample_time_difference,
        genetic_distance,
    )
    ca_profile = compatibility_profile(
        ca_models,
        ca_labels,
        sample_time_difference,
        genetic_distance,
    )
    matrix = compatibility_matrix(sample_time_difference, genetic_distance)
    top_k = min(10, matrix.size)
    top_ranked = ranked_combinations(matrix)[:top_k]

    with output:
        output.clear_output(wait=True)
        fig = plt.figure(figsize=(12, 10), constrained_layout=True)
        grid = fig.add_gridspec(
            2,
            3,
            height_ratios=[1.0, 1.75],
            width_ratios=[1.5, 1.0, 1.5]
        )
        ax_ad = fig.add_subplot(grid[0, 2])
        ax_ca = fig.add_subplot(grid[0, :2])
        ax_rank = fig.add_subplot(grid[1, 0])
        ax_heatmap = fig.add_subplot(grid[1, 1:])

        ad_positions = np.arange(len(ad_labels))
        ca_positions = np.arange(len(ca_labels))

        ax_ad.plot(ad_positions, ad_profile, color="#2a6f97", marker="o", linewidth=2)
        ax_ad.fill_between(ad_positions, ad_profile, color="#2a6f97", alpha=0.12)
        ax_ad.set_xticks(ad_positions, labels=ad_labels)
        ax_ad.set_ylim(0.0, 1.05)
        ax_ad.set(
            title="Single-scenario AD compatibility",
            xlabel="Ancestor-descendant scenario",
            ylabel="Compatibility",
        )
        ax_ad.grid(axis="y", alpha=0.25)

        ax_ca.plot(ca_positions, ca_profile, color="#c84c09", marker="o", linewidth=2)
        ax_ca.fill_between(ca_positions, ca_profile, color="#c84c09", alpha=0.12)
        ax_ca.set_xticks(ca_positions, labels=ca_labels, rotation=35, ha="right")
        ax_ca.set_ylim(0.0, 1.05)
        ax_ca.set(
            title="Single-scenario CA compatibility",
            xlabel="Common-ancestor scenario",
            ylabel="Compatibility",
        )
        ax_ca.grid(axis="y", alpha=0.25)
        ax_ca.tick_params(axis="x", labelsize=8)

        rank_labels = [f"{ad_label} + {ca_label}" for _, ca_label, ad_label in top_ranked]
        rank_scores = [score for score, _, _ in top_ranked]
        bar_colors = plt.cm.magma(np.clip(np.asarray(rank_scores) / 2.0, 0.0, 1.0))
        ax_rank.barh(rank_labels, rank_scores, color=bar_colors)
        ax_rank.invert_yaxis()
        ax_rank.set_xlim(0.0, 1.05)
        ax_rank.set(
            title=f"Top {top_k} scenario pairs",
            xlabel="Compatibility",
        )
        ax_rank.grid(axis="x", alpha=0.25)
        ax_rank.tick_params(axis="y", labelsize=8)
        for rank_index, score in enumerate(rank_scores):
            ax_rank.text(
                min(score + 0.03, 2.01),
                rank_index,
                f"{score:.2f}",
                va="center",
                ha="left",
                fontsize=8,
            )

        image = ax_heatmap.imshow(
            matrix,
            aspect="auto",
            cmap="magma",
            vmin=0.0,
            vmax=1.05,
        )
        for row_index, ca_label in enumerate(ca_labels):
            for col_index, ad_label in enumerate(ad_labels):
                value = matrix[row_index, col_index]
                text_color = "white" if value < 1.1 else "black"
                ax_heatmap.text(
                    col_index,
                    row_index,
                    f"{value:.2f}",
                    ha="center",
                    va="center",
                    color=text_color,
                    fontsize=9,
                )

        ax_heatmap.set_xticks(np.arange(len(ad_labels)), labels=ad_labels, rotation=30, ha="right")
        ax_heatmap.set_yticks(np.arange(len(ca_labels)), labels=ca_labels)
        ax_heatmap.set(
            title="Two-scenario subset compatibility",
            xlabel="Ancestor-descendant scenario",
            ylabel="Common-ancestor scenario",
        )
        ax_heatmap.tick_params(axis="x", labelsize=9)
        ax_heatmap.tick_params(axis="y", labelsize=9)
        ax_heatmap.set_xticks(np.arange(-0.5, len(ad_labels), 1), minor=True)
        ax_heatmap.set_yticks(np.arange(-0.5, len(ca_labels), 1), minor=True)
        ax_heatmap.grid(False)
        ax_heatmap.grid(which="minor", color="white", linewidth=0.8, alpha=0.35)
        ax_heatmap.tick_params(which="minor", bottom=False, left=False)
        fig.colorbar(
            image,
            ax=ax_heatmap,
            shrink=0.92,
            label="Compatibility of [AD(m), CA(mi,mj)]",
        )
        fig.suptitle(
            f"$t_{{i,j}}$ = {sample_time_difference} days, $g_{{i,j}}$ = {genetic_distance}",
            fontsize=16,
            fontweight="bold",
        )
        plt.show()


for control in (time_slider, genetic_slider):
    control.observe(draw_heatmap, names="value")

display(widgets.VBox([time_slider, genetic_slider, output]))
draw_heatmap()
```
